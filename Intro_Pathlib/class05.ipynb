# Introdución a pathlib
## Clase #5: Extraer archivos ZIP

```python
from pathlib import Path #Para manejar rutas de archivos
import zipfile # paraa manejar archivos zip
```

```python
ruta_descomprimido = Path("C1_descomprimido") # Carpeta donde se descomprimirán los archivos
ruta_descomprimido.mkdir(exist_ok=True) # Crear la carpeta si no existe
```

```python
comprimidos = list(Path("D:/Data_Science/Python/Python_teoria_y_ejemplos/Intro_Pathlib").glob("*.zip")) # Lista de archivos zip en la carpeta actual
print(comprimidos)
```

```python
for ruta in comprimidos: # Iterar sobre cada archivo zip en la lista
    print(ruta)
```

```python
for ruta in comprimidos:
    with zipfile.ZipFile(ruta, "r") as zip_object: # Abrir el archivo zip en modo lectura
        zip_object.extractall(path= ruta_descomprimido) # Extraer todo el contenido en la carpeta de descompresión
```
