# Introdución a pathlib
## Clase #2: Manipulación directorios

```python
from pathlib  import Path
import calendar
```

Creación de un directorio

```python
# creamos un objeto path que representa una nueva carpeta
ruta = Path("Nueva_Carpeta")

#Creamos la carpeta en el sistema de archivos
ruta.mkdir(exist_ok=True)  # exist_ok=True evita error si la carpeta ya existe
```

```python
ruta_2 = Path("Nueva_Carpeta_2")
ruta_2.mkdir(exist_ok=True)
```

Creación de una estructura de carpetas

```python
#estructuta de carpetas
ruta_3 = Path("Nueva_carpeta_3/subcarpeta")
ruta_3.mkdir(parents=True, exist_ok=True) # parents=True crea todas las carpetas intermedias necesarias
```

```python
ruta_4 = Path("Nueva_carpeta_4/subcarpeta/subcarpeta_2")
ruta_4.mkdir(parents=True, exist_ok=True)
```

Estructra más compleja usando la librería calendar para obtener los nombres de los meses

```python
months = list(calendar.month_name)[1:]
print(months)
print(type(months))
```

```python
days = [i for i in range(1,31)]
print(days)
```

```python
#Creamos la estructura de carpeta de cada día para cada mese del año
for month in months:
    for day in days:
        ruta_day =Path(f"Año/{month}/{day}")
        ruta_day.mkdir(parents=True,exist_ok=True)
```
