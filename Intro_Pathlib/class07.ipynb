```python
import pandas as pd
from pathlib import Path
```

```python
# Definir la ruta al archivo CSV utilizando Pathlib
file_path = Path(r"ejemplo_practico\datos_crudos\Spotify_Song_Attributes.csv")
file_path
```

```python
# Leer el archivo CSV con pandas
df  = pd.read_csv(file_path)
#mostramos el contenido del DataFrame
df
```

```python
# obtener los generos únicos
generos = df["genre"].unique()
print(f"Total de generos únicos: {len(generos)}\n")
print(generos)
```

```python
# Para cada género creamos un subDataFramey lo guardamos en la carpeta correspondiente
for genero in generos:
    genero_df = df[df["genre"] == genero]
    dir_path = Path(f"ejemplo_practico/generos/{genero}")
    dir_path.mkdir(parents=True, exist_ok=True)
    genero_df.to_csv(dir_path/f"Spotify_Song_Atributes_{genero}_.csv", index=False)
```
