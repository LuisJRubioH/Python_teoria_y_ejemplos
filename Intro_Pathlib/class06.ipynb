## Clase #6: Buscar archivos 

```python
from pathlib import Path
```

```python
objetivo = "archivo_1" 
```

```python
ruta = Path("C1_descomprimido")
```

```python
archivos = list(ruta.glob("*/*.txt")) # Lista de todos los archivos .txt en subcarpetas
print(archivos)
```

```python
for ruta in archivos:
    if objetivo in ruta.name: # Verificar si el nombre del archivo contiene el objetivo
        print(ruta.absolute()) # Imprimir la ruta absoluta del archivo encontrado
```
