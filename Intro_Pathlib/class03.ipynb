# Introdución a pathlib
## Clase #3: Renombrar archivos y directorios

```python
from pathlib import Path
```

Vamos a renombrar dierectorios

```python
ruta = Path("directorio_de_prueba_renombrado")
ruta.rename("C1") #se renombra la carpeta
```

renombramos archivos

```python
ruta_2 = Path("C1/docuento de prueba .txt")
ruta_2.rename("C1/docuento_de_prueba_renombrado.txt") #se renombra la carpeta
```

### Listar directoios y subdirectorios

```python
# Creamos la nueva ruta
carpeta = Path("Año")

#Iteramos sobre todos los subdirectorios de la carpeta intro_pathlib
for subcarpeta in carpeta.iterdir():
    print(subcarpeta)
```

```python
carpeta_2 = Path("D:/Data_Science/Python/Python_teoria_y_ejemplos/Intro_Pathlib").glob("**/*")
for subcarpeta_2 in carpeta_2:
    print(subcarpeta_2)
```

### Imprimir solo las rutas que contingan archivos

```python
carpeta_3 = Path("D:/Data_Science/Python/Python_teoria_y_ejemplos/Intro_Pathlib").glob("**/*")

for subcarpeta in carpeta_3:
    if subcarpeta.is_file():
        print(subcarpeta)
```

### Encontrar la ruta de archivos con una extensión específica

```python
carpeta_4 = Path("D:/Data_Science/Python/Python_teoria_y_ejemplos/Intro_Pathlib").glob("**/*.txt") #busca solo archivos txt

for subcarpeta in carpeta_4:
    print(subcarpeta.name)
```
