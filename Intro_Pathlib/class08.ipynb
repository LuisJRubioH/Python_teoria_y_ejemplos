# Introdución a pathlib
## Clase #8: Cambiar Extensiones de Archivos 

```python
from pathlib import Path
import zipfile
```

```python
ruta_nuevo_formato = Path("archivos_con_formato_cambiado")
ruta_nuevo_formato.mkdir(exist_ok = True)
```

```python
comprimidos =  list(Path("D:/Data_Science/Python/Python_teoria_y_ejemplos/Intro_Pathlib").glob("*.zip")) 
print(comprimidos)
```

```python
# descomprimimos la carpeta C1_comprimido y los guadamos en la carpeta archivos_con_formato_cambiado
for archivo in comprimidos:
    with zipfile.ZipFile(archivo, "r") as zip_object: # Abrir el archivo zip en modo lectura
        zip_object.extractall(path= ruta_nuevo_formato) # Extraer todo el contenido en la carpeta de descompresión
```

```python
for archivo in ruta_nuevo_formato.iterdir():
    print(archivo)
```

```python
# para cambiarles el formato delos archivos de la carpeta arpeta archivos_con_formato_cambiado

for archivo in ruta_nuevo_formato.glob("**/*.txt"):
        nueva_extension = archivo.with_suffix(".csv")
        archivo.rename(nueva_extension)
```
