# Introdución a pathlib
## Clase #4: Crear y eliminar archivos de manera eficiente

```python
from pathlib import Path
import random
```

#### Creación de archivos

```python
random.seed(42) # Fijar la semilla para reproducibilidad
numeros = list(random.randint(1, 100) for _ in range(20)) # Generar una lista de 10 números aleatorios entre 1 y 100
print(numeros)
```

```python
for num in numeros: # Iterar sobre cada número en la lista
    with open(f"c1/archivo_{num}.txt", "w") as file: # Crear y abrir un archivo con el número en su nombre
        file.write("HJola mundo, estamos creando archivos, este es el archivo con número aleatorio: " + str(num))# Escribir contenido en el archivo
```

#### Eliminar archivos

```python
ruta_eliminar = list(Path("C1").glob("**/*.tex"))
for archivo in ruta_eliminar:
    archivo.unlink()
```
