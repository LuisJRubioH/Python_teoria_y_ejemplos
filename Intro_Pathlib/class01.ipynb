Creación de un objeto path

```python
#Creaciónn de un objeto Path a partir de una ruta específica
path_obj1 = Path(r"D:\Data_Science\Python\Python_teoria_y_ejemplos\Intro_Pathlib\directorio_de_prueba\docuento de prueba .txt")
print(path_obj1.parts)
```

```python
#Algunos método inetersantes
print(f"nombre del archivo con extensión: {path_obj1.name}") # Devuelve el nombre del archivo con extension
print(f"nombre del archivo sin extensión: {path_obj1.stem}") # devuelve el nombre del archivo sin extension
print(f"extensión del aerchivo: {path_obj1.suffix}") # Devuelve la extensión del archivo
```
