```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = x**3
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color = "r")
plt.subplot(1,2,2)
plt.hist(y, color = "g")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, color = "g")
plt.subplot(2,1,2)
plt.bar(x, y, color = "b")
plt.show()
```
