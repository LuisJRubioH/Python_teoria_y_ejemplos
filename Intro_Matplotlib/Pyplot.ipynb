```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5, 11)
y = x**2

plt.plot(x,y)
plt.show()
```

Podemos realizar modificaciones de estuilo 

```python

plt.plot(x, y, color ="r", marker = "o", linestyle = "dashed", linewidth = 2, markersize = 7)
plt.show()
```
