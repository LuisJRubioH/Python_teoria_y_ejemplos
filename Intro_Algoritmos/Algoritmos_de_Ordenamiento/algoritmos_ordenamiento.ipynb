### **Ordenamiento de la burbuja (Bubble sort)**

```python
def bubble_sort(arr:list):
    
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                print(f"iteració = {i}\n j = {j} y j+1 = {j+1}\n Swapped {arr[j+1]} and {arr[j]}: {arr}")
    return arr

arr = [64, 34, 25, 100, 12, 22, 11, 90]
print(arr)
print(bubble_sort(arr))
```

### **Ordenamiento de selección (Selection Sort)**

```python
def selection_sort(arr:list):
    
    # recorremos todo nuestro array
    for i in range(len(arr)):
        print(arr)
        # buscamos el minimo en el array no ordenado
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        print(f"min_idx = {min_idx}")
        print(f"Intercambiando {arr[i]} con {arr[min_idx]}")
        
        
        # intercambiamos el minimo encontrado con el primer elemento
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr


arr = [64, 34, 25, 100, 12, 22, -11, 90]
print("Array original:", arr)
sorted_arr = selection_sort(arr)
print("Sorted array:", sorted_arr)
```
